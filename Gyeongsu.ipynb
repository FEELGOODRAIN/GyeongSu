```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.font_manager as fm
import platform
from matplotlib import font_manager , rc

if platform.system() == 'Darwin':
  rc('font' , family = 'AppleGothic')
elif platform.system() == 'Windows':
  path = 'C:/Windows/Fonts/malgun.ttf'
  font_name = font_manager.FontProperties(fname = path).get_name()
  rc('font' , family = font_name)
else:
  print('모름')
import warnings
warnings.filterwarnings('ignore')
```

```python
# 장애인주차구역 위반 건수( 2015~~ 2019) 만들기
x = pd.DataFrame({'년도' : [2015,2016,2017,2018,2019] , '건수' : [152856 , 262068 , 328237 , 414409 , 601513]})
x
```

```python
x.set_index(x['년도']).drop('년도' , axis = 1)
```

```python
plt.plot(np.arange(5) , x['건수'] , marker='o' , color = 'red')

for i in range(len(x)):
    height = x['건수'][i]
    plt.text(i, height + 100 , x['건 수'][i] ,  ha='center', va='bottom', size = 12)

plt.xticks(np.arange(5) , ['2015년','2016년','2017년','2018년','2019년'])

plt.xlabel('년도')
plt.ylabel('적발건수' , rotation = 0)
plt.tick_params(axis="x", labelrotation=45)
plt.title('2015년~2019년 장애인주차구역 위반 건수')
plt.show()
```
