```python
# csv파일 로드
type = pd.read_csv(path + '/2018_장애인유형.csv')
```

```python
# 데이터셋 확인
type
```

```python
# 사용하지 않는 컬럼 제거
del type['Unnamed: 0']
del type['ID']
del type['BR_YY']
del type['ADM_DR_NM']
del type['geometry']
del type['MN_OT']
del type['WN_OT']
```

```python
# 남은 컬럼 확인
type.columns
```

```python
# 데이터 자료형 확인
type.info()
```

```python
# 행정동까지 데이터셋에 있었으므로 자치구별 그룹화
type = type.groupby(by = 'GU_NAM').sum()
type
```

```python
# 자치구, 청각(시각)남자, 청각(시각)여자 컬럼 분리 위해 인덱스 생성
type = type.reset_index()
type
```

```python
# 청각, 시각 따로 빼고 나머지 장애유형 모두 더하기
vd = type[['GU_NAM', 'MN_VD', 'WN_VD']]
hi = type[['GU_NAM', 'MN_HI', 'WN_HI']]
```

```python
# 시각장애인 남, 녀 함산 추출
vd['vdtot'] = vd['MN_VD'] + vd['WN_VD']
vd
```

```python
# 청각장애인 남 녀 합산 추출
hi['hitot'] = hi['MN_HI'] + hi['WN_HI']
hi
```

```python
# type 데이터프레임 복사
typ = type.copy()
```

```python
# type 데이터프레임의 시각, 청각 장애인현황 컬럼 삭제
del typ['MN_VD']
del typ['WN_VD']
del typ['MN_HI']
del typ['WN_HI']
typ
```

```python
# 자치구별 전체 장애인 수 컬럼과 자치구명만 가지고있는 데이터프레임 생성
ty = type[['GU_NAM', 'OT_ALL']]
ty
```

```python
# 자치구별  시각장애인 수, 청각장애인 수, 전체장애인 수 데이터프레임 결합
result = pd.merge(vd, hi, on = 'GU_NAM')
result = pd.merge(result, ty, on = 'GU_NAM')
result
```

```python
# 시각, 청각장애인 비율 컬럼 생성
result['vd'] = round(result['vdtot'] / result['OT_ALL'],4) * 100
result['hi'] = round(result['hitot'] / result['OT_ALL'],4) * 100
result
```

```python
# 그래프화
plt.figure(figsize = (20,8))
bar_width = 0.25
plt.bar(result.index , result['vd'] , width = 0.25, label = '시각장애인 비율')
plt.bar(result.index + bar_width , result['hi'] , width = 0.25, label = '청각장애인 비율')

plt.xticks(result.index , result['GU_NAM'])


plt.ylabel('비율(%)', size = 13)
plt.title('자치구별 장애인 중 시각/청각장애인 비율')
plt.legend()
plt.show()
```
