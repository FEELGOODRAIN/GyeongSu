```python
# csv파일 로드
type = pd.read_csv(path + '/2018_장애인유형.csv')
```

```python
# 데이터셋 확인
type
```

```python
# 사용하지 않는 컬럼 제거
del type['Unnamed: 0']
del type['ID']
del type['BR_YY']
del type['ADM_DR_NM']
del type['geometry']
del type['MN_OT']
del type['WN_OT']
```

```python
# 남은 컬럼 확인
type.columns
```

```python
# 데이터 자료형 확인
type.info()
```

```python
# 행정동까지 데이터셋에 있었으므로 자치구별 그룹화
type = type.groupby(by = 'GU_NAM').sum()
type
```

```python
# 자치구, 청각(시각)남자, 청각(시각)여자 컬럼 분리 위해 인덱스 생성
type = type.reset_index()
type
```

```python
# 청각, 시각 따로 빼고 나머지 장애유형 모두 더하기
vd = type[['GU_NAM', 'MN_VD', 'WN_VD']]
hi = type[['GU_NAM', 'MN_HI', 'WN_HI']]
```

```python
# 시각장애인 남, 녀 함산 추출
vd['vdtot'] = vd['MN_VD'] + vd['WN_VD']
vd
```

```python
# 청각장애인 남 녀 합산 추출
hi['hitot'] = hi['MN_HI'] + hi['WN_HI']
hi
```

```python
# type 데이터프레임 복사
typ = type.copy()
```

```python
# type 데이터프레임의 시각, 청각 장애인현황 컬럼 삭제
del typ['MN_VD']
del typ['WN_VD']
del typ['MN_HI']
del typ['WN_HI']
typ
```

```python
# 자치구별 전체 장애인 수 컬럼과 자치구명만 가지고있는 데이터프레임 생성
ty = type[['GU_NAM', 'OT_ALL']]
ty
```

```python
# 자치구별  시각장애인 수, 청각장애인 수, 전체장애인 수 데이터프레임 결합
result = pd.merge(vd, hi, on = 'GU_NAM')
result = pd.merge(result, ty, on = 'GU_NAM')
result
```

```python
# 시각, 청각장애인 비율 컬럼 생성
result['vd'] = round(result['vdtot'] / result['OT_ALL'],4) * 100
result['hi'] = round(result['hitot'] / result['OT_ALL'],4) * 100
result
```

```python


plt.figure(figsize = (8, 6))

# 그래프화
plt.barh(np.arange(25), result['vd'], barwidth = 0.5, label = '시각장애인 비율')
plt.barh(np.arange(25) + 0.5 , result['hi'], barwidth = 0.5, label = '청각장애인 비율')

plt.xlabel('비율')
plt.ylabel('자치구')
plt.title('자치구별 시각/청각장애인 비율')


plt.show()
```

```python
import pandas as pd

a = [135, 117, 91, 86]
b = [112, 133, 151, 162]
c = [96, 108, 99, 104]
year = ['2018', '2019', '2020', '2021']

df = pd.DataFrame({'shop A' : a, 'shop B' : b, 'shop C' : c}, index = year)
```

```python
# 그림 사이즈, 바 굵기 조정
fig, ax = plt.subplots(figsize=(12,6))
bar_width = 0.25

# 연도가 4개이므로 0, 1, 2, 3 위치를 기준으로 삼음
index = np.arange(4)



# 각 연도별로 3개 샵의 bar를 순서대로 나타내는 과정, 각 그래프는 0.25의 간격을 두고 그려짐
b1 = plt.barh(index, df['shop A'], bar_width, alpha=0.4, color='red', label='shop A')

b2 = plt.barh(index + bar_width, df['shop B'], bar_width, alpha=0.4, color='blue', label='shop B')


# x축 위치를 정 가운데로 조정하고 x축의 텍스트를 year 정보와 매칭
plt.xticks(np.arange(bar_width, 4 + bar_width, 1), year)

# x축, y축 이름 및 범례 설정
plt.xlabel('year', size = 13)
plt.ylabel('revenue', size = 13)
plt.legend()
plt.show()
```
